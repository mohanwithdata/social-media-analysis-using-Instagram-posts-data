```python
import pandas as pd
```

```python
df = pd.read_csv('instagram_data.csv')
df.head(3)
```

```python
df.info()
```

```python
df = df[['owner_username', 'caption', 'comments', 'likes', 'followers', 'following', 'created_at', 'is_video']]
df.head(3)
```

```python
len(df['owner_username'].unique())
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], unit='s')
df['hour'] = df['created_at'].dt.hour
df['day_of_week'] = df['created_at'].dt.day_name()
df.head(3)
```

```python
df.to_csv("instagram_cleaned.csv", index=False)
```
